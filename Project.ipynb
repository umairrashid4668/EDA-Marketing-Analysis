```python
import pandas as pd

# Assuming df is your DataFrame
df['Income'] = df['Income'].replace({'\$': '', ',': ''}, regex=True).str.strip().astype(float)

# Verify the changes
print(df.dtypes)
```

```python


# Assuming df is your DataFrame and 'column_name' is the column you want to impute
mean_value = df['Income'].mean()
df['Income'] = df['Income'].fillna(mean_value)

# Verify the changes
print(df['Income'])
```

```python
df['Income'].isnull().sum()
```

```python

# Assuming df is your DataFrame and you want to combine Columns
df['Revenue_generated'] = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

# Verify the changes
df.head()
```

```python
# Made a colume 'Total_Purchases' by adding all the purchases features

df['Revenue_Purchases'] = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)
df.head()
```

```python

from datetime import datetime

# Assuming df is your DataFrame and it has a 'Year_Birth' column
current_year = datetime.now().year

# Create 'Age' column
df['Age'] = current_year - df['Year_Birth']

# Verify the changes
df.head()
```

```python
# Variable TotalAmount_Spent for each customer by adding all the Mnt* features
df['TotalAmount_Spent'] = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)
df.head(10)
```

```python
import pandas as pd

# Assuming df is your DataFrame and 'Education' is the column you want to map
education_to_years = {
    'Basic': 9,          # Example: Basic education might be 9 years
    '2n Cycle': 12,      # Example: 2nd Cycle might be 12 years
    'Graduation': 16,    # Example: Graduation might be 16 years
    'Master': 18,        # Example: Master might be 18 years
    'PhD': 21            # Example: PhD might be 21 years
}

# Create a new column 'Educational_Years' based on the mapping
df['Educational_Years'] = df['Education'].map(education_to_years)

# Verify the changes
df.head()
```

```python
# A column 'TotalCampaignsAcc' as the total acceptance of advertising campaigns

df['TotalCampaignAcc'] = df[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4','AcceptedCmp5']].sum(axis=1)
df.tail()
```

```python
# Count the number of customers who accepted the last campaign by country
country_success = df[df['Response'] == 1]['Country'].value_counts()

# Plotting the number of customers by country who accepted the last campaign
plt.figure(figsize=(12, 6))
sns.barplot(x=country_success.index, y=country_success.values, palette='cubehelix')
plt.title('Number of Customers Who Accepted the Last Campaign by Country')
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sum the amounts spent on each product
product_totals = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum()

# Add the total amount spent overall
product_totals['TotalAmount_Spent'] = df['TotalAmount_Spent'].sum()

# Plot the data
product_totals.plot(kind='bar', figsize=(12, 8))
plt.title('Total Amount Spent on Each Product and Overall')
plt.xlabel('Products')
plt.ylabel('Total Amount Spent')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Age'].unique()
```

```python

# Define age bins and labels
bins = [0, 18, 30, 45, 60, 100]
labels = ['0-18', '19-30', '31-45', '46-60', '61+']

# Create age group column
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

df.head()
```

```python

# Group by age group and calculate the average amount spent
avg_amount_spent = df.groupby('age_group')['TotalAmount_Spent'].mean()

# Plot the data
avg_amount_spent.plot(kind='bar', figsize=(10, 6))
plt.title('Average Amount Spent by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Amount Spent')
plt.xticks(rotation=45)
plt.show()
```

```python

# Sum the revenue for each product
product_revenue = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum()

# Plot the data
product_revenue.plot(kind='bar', figsize=(12, 8))
plt.title('Total Revenue Generated by Each Product')
plt.xlabel('Products')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python

# Step 2: Melt the DataFrame
# Assuming the campaign columns are named 'cAcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5'
campaign_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
melted_df = df.melt(id_vars=['Country'], value_vars=campaign_columns, var_name='campaign_columns', value_name='accepted')

# Step 3: Filter the DataFrame
accepted_df = melted_df[melted_df['accepted'] == 1]

# Step 4: Group by country and campaign, and count
country_campaign_counts = accepted_df.groupby(['Country', 'campaign_columns']).size().reset_index(name='counts')

# Step 5: Plot the results
plt.figure(figsize=(14, 8))
sns.barplot(x='Country', y='counts', hue='campaign_columns', data=country_campaign_counts)
plt.title('Number of Campaign Acceptances by Country and Campaign')
plt.xlabel('Country')
plt.ylabel('Number of Acceptances')
plt.xticks(rotation=45)
plt.legend(title='Campaign')
plt.show()
```
